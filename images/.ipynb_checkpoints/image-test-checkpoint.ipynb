```python
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
%matplotlib inline
animals_mask = np.array(Image.open("Animals.png").convert('P'))
animals_mask
```

```python
arts_mask = np.array(Image.open("Arts, Culture, Humanities.png").convert('P'))
for i in range(len(arts_mask)):
    print(arts_mask[i])
```

```python
def transform_format(val):
    if val == 0:
        return 255
    else:
        return val

x = []
transformed_mask = np.zeros((animals_mask.shape[0], animals_mask.shape[1]), dtype=np.int32)
transformed_arts_mask = np.zeros((arts_mask.shape[0], arts_mask.shape[1]), dtype=np.int32)

for i in range(len(animals_mask)):
    transformed_mask[i] = list(map(transform_format, animals_mask[i]))
for i in range(len(arts_mask)):
    transformed_arts_mask[i] = list(map(transform_format, arts_mask[i]))
transformed_arts_mask
```

```python
from wordcloud import WordCloud
wc = WordCloud(background_color="white", mask=transformed_arts_mask, contour_width=3, contour_color='black', max_words=2000, width=800, height=400)
wc.generate('yes yes yes yes no no no why why please please yes yes yes')
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
